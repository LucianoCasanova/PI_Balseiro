```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def leer_datos(file_path,start_line):
    
    valores = []
    
    with open(file_path, 'r') as file:
        
        lines = file.readlines()
        
        for line in lines[start_line:]:
            
            parts = line.split()
            
            if len(parts) >= 2:
                
                valor = float(parts[1])
                valores.append(valor)
    
    return np.array(valores)
```

```python
rho = 997
```

```python
ivc_p_path = 'patchAverage(total(p),name=IVC,patch=IVC)/0/surfaceFieldValue.dat'
svc_p_path = 'patchAverage(total(p),name=SVC,patch=SVC)/0/surfaceFieldValue.dat'
lpa_p_path = 'patchAverage(total(p),name=LPA,patch=LPA)/0/surfaceFieldValue.dat'
rpa_p_path = 'patchAverage(total(p),name=RPA,patch=RPA)/0/surfaceFieldValue.dat'

ivc_p = rho*leer_datos(ivc_p_path, 4)
svc_p = rho*leer_datos(svc_p_path, 4)
lpa_p = rho*leer_datos(lpa_p_path, 4)
rpa_p = rho*leer_datos(rpa_p_path, 4)

ivc_q_path = 'patchFlowRate(patch=IVC)/0/surfaceFieldValue.dat'
svc_q_path = 'patchFlowRate(patch=SVC)/0/surfaceFieldValue.dat'
lpa_q_path = 'patchFlowRate(patch=LPA)/0/surfaceFieldValue.dat'
rpa_q_path = 'patchFlowRate(patch=RPA)/0/surfaceFieldValue.dat'

#Agrego un 0 que no se por que no me lo da OpenFoam

ivc_q = -1*np.insert(leer_datos(ivc_q_path,4),0,0)
svc_q = -1*np.insert(leer_datos(svc_q_path,4),0,0)
lpa_q = np.insert(leer_datos(lpa_q_path,4),0,0)
rpa_q = np.insert(leer_datos(rpa_q_path,4),0,0)
```

```python
len(ivc_q), len(ivc_p)
```

```python
n=0
t = np.arange(n*5e-2, 7.55, 0.05)

%matplotlib inline

fig, axs = plt.subplots(4, 1, figsize=(10, 12))

axs[0].plot(t, ivc_p[n:], color='b')
axs[0].set_title('IVC Ptotal vs Time')
axs[0].set_ylabel('IVC Ptotal')

axs[1].plot(t, svc_p[n:], color='r')
axs[1].set_title('SVC Ptotal vs Time')
axs[1].set_ylabel('SVC Ptotal')

axs[2].plot(t, lpa_p[n:], color='g')
axs[2].set_title('LPA Ptotal vs Time')
axs[2].set_ylabel('LPA Ptotal')

axs[3].plot(t, rpa_p[n:], color='k')
axs[3].set_title('RPA Ptotal vs Time')
axs[3].set_ylabel('RPA Ptotal')

for ax in axs:
    ax.set_xlabel('Time (t)')

plt.tight_layout()

plt.show()
```

```python
n=0
t = np.arange(n*5e-2, 7.55, 0.05)

fig, axs = plt.subplots(4, 1, figsize=(10, 12))

axs[0].plot(t, ivc_q[n:], color='b')
axs[0].set_title('IVC Flow vs Time')
axs[0].set_ylabel('IVC Flow')

axs[1].plot(t, svc_q[n:], color='r')
axs[1].set_title('SVC Flow vs Time')
axs[1].set_ylabel('SVC Flow')

axs[2].plot(t, lpa_q[n:], color='g')
axs[2].set_title('LPA Flow vs Time')
axs[2].set_ylabel('LPA Flow')

axs[3].plot(t, rpa_q[n:], color='k')
axs[3].set_title('RPA Flow vs Time')
axs[3].set_ylabel('RPA Flow')

for ax in axs:
    ax.set_xlabel('Time (t)')

plt.tight_layout()

plt.show()
```

```python
ivc_power = ivc_p*ivc_q
svc_power = svc_p*svc_q
lpa_power = lpa_p*lpa_q
rpa_power = rpa_p*rpa_q
```

```python
n=0
t = np.arange(n*5e-2, 7.55, 0.05)

fig, axs = plt.subplots(4, 1, figsize=(10, 12))

axs[0].plot(t, ivc_power[n:], color='b')
axs[0].set_title('IVC Power vs Time')
axs[0].set_ylabel('IVC Power')

axs[1].plot(t, svc_power[n:], color='r')
axs[1].set_title('SVC Power vs Time')
axs[1].set_ylabel('SVC Power')

axs[2].plot(t, lpa_power[n:], color='g')
axs[2].set_title('LPA Power vs Time')
axs[2].set_ylabel('LPA Power')

axs[3].plot(t, rpa_power[n:], color='k')
axs[3].set_title('RPA Power vs Time')
axs[3].set_ylabel('RPA Power')

for ax in axs:
    ax.set_xlabel('Time (t)')

plt.tight_layout()

plt.show()
```
